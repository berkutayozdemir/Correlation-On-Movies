```python
# importing libraries

import pandas as pd 
import numpy as np 
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt 

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusting the plot diameters to 12 and 8 
```

```python
# Read and import the data into notebook
pd.read_csv("movies.csv")
df = pd.read_csv("movies.csv")
df.dropna()
```

```python
df.head()  # Shows the first 5 rows of the dataframe
```

```python
# Missing value analysis 

df.dropna()
for col in df.columns :
    missng_percentage = np.mean(df[col].isnull())
    print('{} - {}%'.format (col,missng_percentage)) # Format the data --> Column - Corresponding missing percentage 
```

```python
# Finding the data types of the columns

df.dtypes
```

```python
# Changing the data types of columns
df = df.dropna(subset=["budget"])
df["budget"] = df["budget"].astype("int64")

df = df.dropna(subset=["gross"])
df["gross"] = df["gross"].astype("int64")
```

```python
# Sorting the movies by gross as descending
df.sort_values(by=["gross"] , inplace=False , ascending= False) 
```

```python
# Drop any duplicates 
df["company"].drop_duplicates().sort_values(ascending=False)
```

```python
df
```

## Correlation Analysis

```python
x=df["budget"]
y=df["gross"]
plt.scatter(x, y,alpha=0.8)
plt.title("Budget vs Gross")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
df.sort_values(by= "gross" , ascending= False)
```

```python
sns.regplot(x= "budget" , y= "gross", data=df , line_kws={"color": "Black"}, scatter_kws= {"color" :"Orange" })
```

As we can see on the correlation graph budget and gross are correlated. They have positive linear relationship.

```python
numeric_df = df.select_dtypes(include=[float , int])  # excluded non numerical columns
numeric_df.corr(method= "pearson")
```

```python
ax = sns.heatmap(numeric_df.corr() , annot= True)
plt.title("Correlation Matrix")
plt.show()
```

```python

df_numeric = df
for col in df_numeric.columns:
    if(df_numeric[col].dtype == "object") :
        df_numeric[col] = df_numeric[col].astype("category")
        df_numeric[col]= df_numeric[col].cat.codes
df_numeric.sort_values(by="gross" , ascending= False)


```

Wrote a for loop that converts non numeric objects into a numeric string object. So that we can conduct an analysis about companies correlations between other numerical fields. 

```python
correlation_matrix = df_numeric.corr()
# Unstack the correlation matrix and sort the values
corr_pairs = correlation_matrix.unstack().sort_values(kind="quicksort")

# Filter out self-correlations
filtered_corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

print(filtered_corr_pairs)
```

```python
sns.heatmap(df_numeric.corr() , annot=True)
```

As we can see from the heatmap above, the company has no significant effect on the gross revenue that movies make, nor on the budget. Correlations that are close to zero indicate a low or near-zero relationship between these variables.

```python
high_corr_values = filtered_corr_pairs.sort_values(ascending=False)
high_corr_values
```

```python
high_corr_values = high_corr_values[(high_corr_values > 0.5)]
high_corr_values
```

# Conclusion


The analysis reveals a strong correlation between budget and gross revenue, indicating that movies with higher budgets tend to generate higher gross earnings. This suggests that increased investment in a movie's production is often associated with better financial performance at the box office. Additionally, there is a moderate correlation between the gross revenue and the number of votes a movie receives, implying that more popular movies, as indicated by higher vote counts, tend to have higher gross revenues.However, other variables, such as the company's involvement, genre, director, and runtime, do not exhibit significant correlations with gross revenue or with each other.
